```python
# Make sure the plotting function exists
def maybe_group_stat(df, by_col, value_col, agg="median", top_n=10):
    if by_col in df.columns and value_col in df.columns:
        out = df.groupby(by_col)[value_col].agg(agg).sort_values(ascending=False).head(top_n)
        display(out)
        plt.figure()
        out.plot(kind="bar")
        plt.title(f"{agg.title()} {value_col} by {by_col}")
        plt.ylabel(f"{agg.title()} {value_col}")
        plt.tight_layout()
        plt.show()
    else:
        print(f"Skipped: {by_col} or {value_col} not in DataFrame columns")

# Your list of pairs
pairs = [
    ("industry", "revenue"),
    ("industry", "growth_rate"),
    ("stage", "revenue"),
    ("stage", "funding_requested"),
    ("country", "revenue"),
    ("region", "revenue"),
]

# Loop over pairs to run the analysis
for by, val in pairs:
    maybe_group_stat(df, by, val, agg="median", top_n=8)
```

```python
# Cell 11 â€” Outlier detection (z-score)
from scipy import stats

outlier_report = {}
for col in numeric_cols:
    series = df[col].dropna()
    if len(series) >= 10 and series.std(ddof=0) > 0:
        z = (series - series.mean()) / series.std(ddof=0)
        outliers = series[abs(z) > 3]
        if not outliers.empty:
            outlier_report[col] = outliers.sort_values(ascending=False).head(10)

if outlier_report:
    for col, vals in outlier_report.items():
        print(f"\nPotential outliers in {col} (|z|>3):")
        display(vals)
else:
    print("No strong z-score outliers detected (|z|>3).")
```

```python
import pandas as pd

# Load dataset
df = pd.read_excel("Vester DB.xlsx")

# Clean 'current_annual_revenue_usd' column
if 'current_annual_revenue_usd' in df.columns:
    df['current_annual_revenue_usd'] = (
        df['current_annual_revenue_usd']
        .astype(str)                     # make sure it's string
        .str.replace(r'[^0-9.]', '', regex=True)  # remove USD, commas, any non-numeric
    )
    df['current_annual_revenue_usd'] = pd.to_numeric(df['current_annual_revenue_usd'], errors='coerce')

# Now calculate median revenue by industry
if 'company_industry' in df.columns:
    median_revenue_industry = df.groupby('company_industry')['current_annual_revenue_usd']\
                                 .median().sort_values(ascending=False)
    print("Median Revenue by Industry:")
    print(median_revenue_industry)
else:
    print("Column 'company_industry' not found.")
```

```python
import matplotlib.pyplot as plt

# Median Revenue by Industry
plt.figure(figsize=(10,5))
median_revenue_industry.plot(kind='bar', color='skyblue')
plt.title('Median Revenue by Industry')
plt.ylabel('Median Revenue (USD)')
plt.xlabel('Industry')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Median Revenue by Funding Stage
if 'company_funding_stage' in df.columns:
    median_revenue_stage = df.groupby('company_funding_stage')['current_annual_revenue_usd']\
                             .median().sort_values(ascending=False)
    
    plt.figure(figsize=(10,5))
    median_revenue_stage.plot(kind='bar', color='lightgreen')
    plt.title('Median Revenue by Funding Stage')
    plt.ylabel('Median Revenue (USD)')
    plt.xlabel('Funding Stage')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
else:
    print("Column 'company_funding_stage' not found.")
```
